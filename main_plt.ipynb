```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,3,5,8,9,1,7,11]
plt.plot(x,y,marker='*',color='green')
plt.title("Example")
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
x=['red','orange','blue','pink']
y=[2,15,9,11]
plt.bar(x,y,color='red')
plt.title("Example")
plt.xlabel('color')
plt.ylabel('count')
plt.show()
```

```python
x=['red','orange','blue','pink']
y=[2,15,9,11]
plt.pie(y,labels=x,autopct='%1.1f%%', colors=['red','orange','blue','pink'])
plt.show()
```

```python
import numpy as np
x=np.random.rand(1000,1)
plt.hist(x,bins=10)
plt.show()
```

```python
x=np.random.rand(64,64)
plt.imshow(x, cmap='hot')
plt.colorbar()
plt.show()
```
