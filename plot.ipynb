```python
plt.plot(x, y,'o-g')
plt.plot(x, y2)

plt.savefig('out.png') 
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(['sin','cos'])
plt.title('Some plots')
plt.xlabel('Values')
plt.ylabel('Sin & Cos')
plt.grid()
plt.xlim(0,360)
plt.ylim(-1,1)
# plt.axis('tight')
plt.show();
```

```python

x=np.linspace(1,360,100)
y=np.sin(np.radians(x))
y2=np.log(x)
fig,ax=plt.subplots()
ax.plot(x,y,'g')
ax.set_ylabel('$Sin(x)$')
ax2=ax.twinx()
ax2.plot(x,y2,'r')
ax2.set_ylabel('$Log(x)$')
plt.show()
```
